```python
def sum(n):
    if n == 1:
        return 1
    else:
        result = sum(n-1)
        return n + result

print(sum(4))
    
```

```python
def pwer2(n):
    if n == 0:
        return 1
    elif n == 1:
        return 2
    else:
        result = pwer2(n-1)
        return 2 * result

print(pwer2(3))
    
```

```python
def sum_of_natural_numbers(n):
    """
    Function to calculate the sum of the first n natural numbers using recursion.
    
    Parameters:
    n (int): The non-negative integer for which the sum is to be calculated.
    
    Returns:
    int: The sum of the first n natural numbers.
    """
    # Your code here
    if n == 1 or n == 0 :
        return 1
    result = sum_of_natural_numbers(n-1)
    return n + result
print(sum_of_natural_numbers(1))
```

```python
def num_of_digits(n):
    
    if n // 10 == 0:
        return 1
    
    result = num_of_digits(n//10)
    
    return 1 + result     
    
num_of_digits(12335)
```

```python
def check_sorted(l1):
    if len(l1) <= 1:
        return True
    if l1[0] <= l1[1]:
        result = check_sorted(l1[1:])
        return result
    else:
        return False

check_sorted([1,2,3,1])
```

```python
def get_first_index(arr, x):
    if len(arr) == 0:
        return -1  # not found

    if arr[0] == x:
        return 1  # found at first position

    result = get_first_index(arr[1:], x)

    if result == -1:
        return -1  # not found in rest
    else:
        return result + 1
    
result = get_first_index([1,8,3,4,5,2], 2) - 1 
print("result", result)

```

```python
def get_last_index(arr, x):
    if len(arr) == 0:
        return -1  # not found

    result = get_last_index(arr[1:],x)
    
    if result != -1:
        return result + 1
    if arr[0] == x:
        return 0
    return -1
    

    
result = get_last_index([1,8,2,4,5], 2) 
print("result", result)

```
